```python
import streamlit as st
import PyPDF2
import docx
```

```python
def read_file(file):
    if file.type == "application/pdf":
        reader = PyPDF2.PdfFileReader(file)
        text = ""
        for page in range(reader.getNumPages()):
            text += reader.getPage(page).extract_text()
    elif file.type == "application/vnd.openxmlformats-officedocument.wordprocessingml.document":
        doc = docx.Document(file)
        text = ""
        for para in doc.paragraphs:
            text += para.text
    else:
        text = file.read().decode('utf-8')
    return text

st.title("Resume Experience Extractor")
st.write("This app compares a job description with a resume and extracts the number of years of relevant work experience from the resume.")

# File uploader for job description
job_desc_file = st.file_uploader("Upload Job Description", type=['txt', 'pdf', 'docx'])
resume_file = st.file_uploader("Upload Resume", type=['txt', 'pdf', 'docx'])

if job_desc_file and resume_file:
    job_desc_text = read_file(job_desc_file)
    resume_text = read_file(resume_file)

    st.write("Job Description:")
    st.write(job_desc_text)
    
    st.write("Resume:")
    st.write(resume_text)
```

```python
streamlit run app.py
```
